```python
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

# Custom Dataset
class CharTextDataset(Dataset):
    def __init__(self, texts, labels, char_vocab, max_len=128):
        self.texts = texts
        self.labels = labels
        self.char_vocab = char_vocab
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx]) if pd.notnull(self.texts[idx]) else ""
        label = self.labels[idx]
        # Character-level tokenization
        chars = list(text)[:self.max_len]
        char_ids = [self.char_vocab.get(char, self.char_vocab["<UNK>"]) for char in chars]
        if len(char_ids) < self.max_len:
            char_ids += [self.char_vocab["<PAD>"]] * (self.max_len - len(char_ids))
        else:
            char_ids = char_ids[:self.max_len]
        return {
            "input_ids": torch.tensor(char_ids, dtype=torch.long),
            "labels": torch.tensor(label, dtype=torch.long)
        }

# LSTM Model
class CharLSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, num_labels):
        super(CharLSTMClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True, num_layers=2, dropout=0.2)
        self.fc = nn.Linear(hidden_dim, num_labels)

    def forward(self, input_ids):
        embedded = self.embedding(input_ids)
        _, (hidden, _) = self.lstm(embedded)
        output = self.fc(hidden[-1])
        return output

# Load data
train_df = pd.read_csv("culturaX_train_70.csv")
test_df = pd.read_csv("culturaX_test_30.csv")

# Handle missing text
train_df["text"] = train_df["text"].fillna("").astype(str)
test_df["text"] = test_df["text"].fillna("").astype(str)

# Encode labels
label_encoder = LabelEncoder()
train_df["labels"] = label_encoder.fit_transform(train_df["language"])
test_df["labels"] = label_encoder.transform(test_df["language"])

# Build character vocabulary
all_texts = train_df["text"].tolist() + test_df["text"].tolist()
chars = set()
for text in all_texts:
    if text:
        chars.update(list(text))
char_vocab = {"<PAD>": 0, "<UNK>": 1}
for i, char in enumerate(chars, 2):
    char_vocab[char] = i

# Create datasets
train_dataset = CharTextDataset(train_df["text"].tolist(), train_df["labels"].tolist(), char_vocab)
test_dataset = CharTextDataset(test_df["text"].tolist(), test_df["labels"].tolist(), char_vocab)

# Data loaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32)

# Model parameters
vocab_size = len(char_vocab)
embed_dim = 64
hidden_dim = 128
num_labels = len(label_encoder.classes_)

# Initialize model, loss, optimizer
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CharLSTMClassifier(vocab_size, embed_dim, hidden_dim, num_labels).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 5
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for batch in train_loader:
        input_ids = batch["input_ids"].to(device)
        labels = batch["labels"].to(device)
        optimizer.zero_grad()
        outputs = model(input_ids)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {total_loss/len(train_loader):.4f}")

    # Evaluation
    model.eval()
    preds, true_labels = [], []
    with torch.no_grad():
        for batch in test_loader:
            input_ids = batch["input_ids"].to(device)
            labels = batch["labels"].to(device)
            outputs = model(input_ids)
            preds.extend(outputs.argmax(dim=1).cpu().numpy())
            true_labels.extend(labels.cpu().numpy())
    accuracy = accuracy_score(true_labels, preds)
    f1 = f1_score(true_labels, preds, average="weighted")
    print(f"Accuracy: {accuracy:.4f}, F1-score: {f1:.4f}")

# Save model and metadata
torch.save(model.state_dict(), "language_detection_char_lstm_model.pt")
import joblib
joblib.dump(label_encoder, "label_encoder.pkl")
joblib.dump(char_vocab, "char_vocab.pkl")
```

```python
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, f1_score
import joblib

# Custom Dataset
class CharTextDataset(Dataset):
    def __init__(self, texts, labels, char_vocab, max_len=128):
        self.texts = texts
        self.labels = labels if labels is not None else [0] * len(texts)
        self.char_vocab = char_vocab
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx]) if pd.notnull(self.texts[idx]) else ""
        label = self.labels[idx]
        chars = list(text)[:self.max_len]
        char_ids = [self.char_vocab.get(char, self.char_vocab["<UNK>"]) for char in chars]
        if len(char_ids) < self.max_len:
            char_ids += [self.char_vocab["<PAD>"]] * (self.max_len - len(char_ids))
        else:
            char_ids = char_ids[:self.max_len]
        return {
            "input_ids": torch.tensor(char_ids, dtype=torch.long),
            "labels": torch.tensor(label, dtype=torch.long)
        }

# LSTM Model
class CharLSTMClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, num_labels):
        super(CharLSTMClassifier, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.lstm = nn.LSTM(embed_dim, hidden_dim, batch_first=True, num_layers=2, dropout=0.2)
        self.fc = nn.Linear(hidden_dim, num_labels)

    def forward(self, input_ids):
        embedded = self.embedding(input_ids)
        _, (hidden, _) = self.lstm(embedded)
        output = self.fc(hidden[-1])
        return output

# Load saved model, label encoder, and vocab
model_path = "language_detection_char_lstm_model.pt"
label_encoder_path = "label_encoder.pkl"
vocab_path = "char_vocab.pkl"

char_vocab = joblib.load(vocab_path)
label_encoder = joblib.load(label_encoder_path)
vocab_size = len(char_vocab)
embed_dim = 64
hidden_dim = 128
num_labels = len(label_encoder.classes_)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CharLSTMClassifier(vocab_size, embed_dim, hidden_dim, num_labels).to(device)
model.load_state_dict(torch.load(model_path))
model.eval()

# Load test data
test_df = pd.read_csv("culturaX_test_30.csv")
test_df["text"] = test_df["text"].fillna("").astype(str)
test_df["labels"] = label_encoder.transform(test_df["language"])

# Create test dataset and loader
test_dataset = CharTextDataset(test_df["text"].tolist(), test_df["labels"].tolist(), char_vocab)
test_loader = DataLoader(test_dataset, batch_size=32)

# Evaluate on test data
preds, true_labels = [], []
with torch.no_grad():
    for batch in test_loader:
        input_ids = batch["input_ids"].to(device)
        labels = batch["labels"].to(device)
        outputs = model(input_ids)
        preds.extend(outputs.argmax(dim=1).cpu().numpy())
        true_labels.extend(labels.cpu().numpy())

# Calculate metrics
accuracy = accuracy_score(true_labels, preds)
f1 = f1_score(true_labels, preds, average="weighted")
print(f"Test Accuracy: {accuracy:.4f}, Test F1-score: {f1:.4f}")

# Function to predict language for new text
def predict_language(text, model, tokenizer, char_vocab, label_encoder, max_len=128):
    text = str(text) if pd.notnull(text) else ""
    chars = list(text)[:max_len]
    char_ids = [char_vocab.get(char, char_vocab["<UNK>"]) for char in chars]
    if len(char_ids) < max_len:
        char_ids += [char_vocab["<PAD>"]] * (max_len - len(char_ids))
    input_ids = torch.tensor([char_ids], dtype=torch.long).to(device)

    model.eval()
    with torch.no_grad():
        output = model(input_ids)
        pred_label = output.argmax(dim=1).cpu().numpy()[0]
    return label_encoder.inverse_transform([pred_label])[0]

# Example usage
sample_text = "cos'è 'SC' in Ocrevus"
predicted_language = predict_language(sample_text, model, None, char_vocab, label_encoder)
print(f"Predicted language: {predicted_language}")
```
